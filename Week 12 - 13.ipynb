```python
# Exercise 1 : Identifying and Handling Missing Data
import pandas as pd 

# Sample dataset with missing values
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age' : [24, 30, None, 22, 35],
    'Salary' : [48000, None, 57000, None, 60000]
}

df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning: \n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
# Practice Task Week 12
# - Load a dataset of your choice and identify missing values.
# - Implement data transformations to normalize numerical columns.
# - Standardize categorical columns and remove duplicates.

import pandas as pd

# Sample dataset with missing values
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age' : [24, 30, None, 22, 35],
    'Salary' : [48000, None, 57000, None, 60000],
    'Department' : ['HR', 'IT', 'HR', 'Finance', 'Finance']
}

df = pd.DataFrame(data)

print("Jumlah data yang hilang pada setiap kolom:\n", df.isnull().sum())
```

```python
# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Min-Max Scaling
df['Age-Normalized'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
df['Salary-Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

print("\nAfter normalization:\n", df[['Name', 'Age', 'Age-Normalized', 'Salary', 'Salary-Normalized']])
```

```python
# Standardize text values in the 'Department' column
df['Department'] = df['Department'].str.upper()

# Remove duplicates
df = df.drop_duplicates(subset=['Department'])

print("\nAfter standardizing and removing duplicates:\n", df[['Name', 'Department']])
```

Use Matplotlib to Create a Multi-Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D']
group1 = [20, 34, 30, 35]
group2 = [25, 32, 34, 20]
group3 = [22, 30, 32, 25]

x = np.arange(len(categories))  # Label locations
width = 0.25  # Width of the bars

# Create the multi-bar chart
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - width, group1, width, label='Group 1', color='skyblue')
ax.bar(x, group2, width, label='Group 2', color='orange')
ax.bar(x + width, group3, width, label='Group 3', color='green')

# Add labels, title, and legend
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Comparison of Groups by Categories')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```

Generate Scatter Plots with Seaborn

```python
import seaborn as sns
import pandas as pd

# Sample dataset
data = {
    'Age': [22, 25, 30, 35, 40, 28, 33, 24],
    'Salary': [25000, 27000, 30000, 35000, 40000, 32000, 37000, 26000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Finance', 'HR', 'Finance', 'IT']
}

df = pd.DataFrame(data)

# Create scatter plot
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=df,
    x='Age',
    y='Salary',
    hue='Department',
    style='Department',
    palette='Set2',
    s=100  # Size of markers
)

# Customize the plot
scatter.set_title('Scatter Plot: Age vs Salary by Department')
scatter.set_xlabel('Age')
scatter.set_ylabel('Salary')
plt.legend(title='Department')
plt.show()
```

Customize Plot Layouts and Color Schemes

```python
# Create a customized scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=df,
    x='Age',
    y='Salary',
    hue='Department',
    palette='coolwarm',
    s=120
)

# Add grid, title, and labels
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Customized Scatter Plot: Age vs Salary', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.legend(title='Department', fontsize=10, title_fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Load dataset
df = pd.read_csv('CarsDataset.csv')

print(df)

# 2. Menghapus baris dengan nilai kosong
df_clean = df.dropna().reset_index(drop=True)

# 3. Bersihkan nama kolom dari spasi ekstra
df_clean.columns = df_clean.columns.str.strip()

# Ubah kolom 'Km/h' dan 'Price' dari string ke numerik (hapus koma)
df_clean['Km/h'] = df_clean['Km/h'].str.replace(',', '').astype(float)
df_clean['Price'] = df_clean['Price'].str.replace(',', '').astype(float)

# 4. Deteksi dan penanganan outlier menggunakan IQR
Q1 = df_clean[['Km/h', 'Price']].quantile(0.25)
Q3 = df_clean[['Km/h', 'Price']].quantile(0.75)
IQR = Q3 - Q1

# Menghapus outlier di luar rentang IQR
df_clean = df_clean[~((df_clean[['Km/h', 'Price']] < (Q1 - 1.5 * IQR)) | 
                      (df_clean[['Km/h', 'Price']] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Tampilkan data yang sudah dibersihkan
print("\n",df_clean)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot distribusi harga mobil
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['Price'], bins=30, kde=True)
plt.title('Distribusi Harga Mobil')
plt.xlabel('Harga')
plt.show()

# Scatter plot antara Tahun dan Harga
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='Year', y='Price', hue='Fuel')
plt.title('Harga Mobil Berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Harga')
plt.show()
```
